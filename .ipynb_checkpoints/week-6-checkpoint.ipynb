```python
from urllib.request import urlopen
```

```python
passage = urlopen('http://k2kteaching.com/registration.html')
```

```python
passage_words = []
```

```python
for a in passage:
    a_words = a.split()
    for word in a_words:
        passage_words.append(word)
```

```python
passage.close()
```

```python
passage_words
```

```python
passage = urlopen('http://k2kteaching.com/registration.html')
```

```python
passage_words = []
```

```python
for a in passage:
    a_words = a.decode('utf8').split()
    for word in a_words:
        passage_words.append(word)
```

```python
passage.close()
```

```python
passage_words
```
